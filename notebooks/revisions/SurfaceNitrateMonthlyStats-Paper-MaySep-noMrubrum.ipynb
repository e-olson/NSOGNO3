```python
sourcedir='/data/eolson/results/MEOPAR/SS36runs/CedarRuns/rev_noMrubrum/'
```

```python
recalc=True
if recalc:
    means=dict()
    stds=dict()
    N=dict()
    for imon in range(5,10):
        N[imon]=list()
        for ind,row in filelist.loc[filelist.mon==imon].iterrows():
            with nc.Dataset(sourcedir+row['file']) as fi:
                N[imon].append(fi.variables['nitrate'][row['ts']:row['te'],k,j0:j1,i0:i1])
        N[imon]=np.concatenate(N[imon],axis=0)
        means[imon]=np.mean(N[imon],0)
        stds[imon]=np.std(N[imon],0)
    with open(saveloc+'monthlySurfaceMeansMaySepNoMrubrum.pkl', 'wb') as output:
        pickle.dump(means, output)
    with open(saveloc+'monthlySurfaceStdsMaySepNoMrubrum.pkl', 'wb') as output:
        pickle.dump(stds, output)
else:
    with open(saveloc+'monthlySurfaceMeansMaySepNoMrubrum.pkl', 'rb') as output:
        means=pickle.load(output)
    with open(saveloc+'monthlySurfaceStdsMaySepNoMrubrum.pkl', 'rb') as output:
        stds=pickle.load(output)
```
