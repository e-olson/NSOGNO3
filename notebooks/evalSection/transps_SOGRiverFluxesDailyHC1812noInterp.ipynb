```python
with nc.Dataset('/ocean/eolson/MEOPAR/NEMO-forcing/grid/mesh_mask201702_noLPE.nc') as fm:
    A=fm.variables['e1t'][0,jg0:jg1,ig0:ig1]*fm.variables['e2t'][0,jg0:jg1,ig0:ig1]*fm.variables['tmask'][0,0,jg0:jg1,ig0:ig1]
SOGmask=tmaskSOG[0,0,jg0:jg1,ig0:ig1]
Framask=tmaskFra[jg0:jg1,ig0:ig1]
nh4=nh4[0,jg0:jg1,ig0:ig1]
```

```python
## for HC201812 only need to consider nh4, silicate and no3
```

```python
# calc daily river flux starting at 1am, with interpolation as in model, sum to daily rates
rau0 = 1026.0
t0=dt.datetime(2015,1,1)
te=dt.datetime(2018,1,1)
runlen=int((te-t0).total_seconds()/(24*3600))
fformatbio='/data/eolson/results/MEOPAR/NEMO-forcing-new/rivers-climatology/bio/R201809_2_bioClim_m%md%d.nc'
fformatFlow='/results/forcing/rivers/R201702DFraCElse_y%Ym%md%d.nc'
var=np.zeros((runlen,))
# SOG Rivers includes Fraser, but also save Fraser separately for reference
sumSilSOGRivers=np.copy(var) # silicate
sumSilFraser=np.copy(var)
sumNO3SOGRivers=np.copy(var)
sumNO3Fraser=np.copy(var)
sumNH4SOGRivers=np.copy(var)
sumNH4Fraser=np.copy(var)
sumLivSOGRivers=np.copy(var)
sumLivFraser=np.copy(var)
times=list()
jj=-1
# NEMO performs interpolation between mid-day of neighboring days (assumes forcing data at noon)
for idt0 in [t0+dt.timedelta(days=ii*1) for ii in range(0,runlen)]:
    jj=jj+1
    print(idt0)
    fbion=nc.Dataset((idt0).strftime(fformatbio))
    ffln=nc.Dataset((idt0).strftime(fformatFlow))
    flow=ffln.variables['rorunoff'][0,jg0:jg1,ig0:ig1]
    sil=fbion.variables['sil'][0,jg0:jg1,ig0:ig1]
    no3=fbion.variables['no3'][0,jg0:jg1,ig0:ig1]
    times.append(idt0)
    sumSilSOGRiversx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*SOGmask*flow*sil,1),0) # hourly rates
    sumSilFraserx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*Framask*flow*sil,1),0) # hourly rates
    sumNO3SOGRiversx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*SOGmask*flow*no3,1),0)
    sumNO3Fraserx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*Framask*flow*no3,1),0)
    sumNH4SOGRiversx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*SOGmask*flow*nh4,1),0)
    sumNH4Fraserx=np.sum(np.sum((1/rau0*3600*24*1e-3)*A*Framask*flow*nh4,1),0)
    sumSilSOGRivers[jj]=np.sum(sumSilSOGRiversx) # sum to daily rates (mol/day)
    sumSilFraser[jj]=np.sum(sumSilFraserx) 
    sumNO3SOGRivers[jj]=np.sum(sumNO3SOGRiversx)
    sumNO3Fraser[jj]=np.sum(sumNO3Fraserx)
    sumNH4SOGRivers[jj]=np.sum(sumNH4SOGRiversx)
    sumNH4Fraser[jj]=np.sum(sumNH4Fraserx)
    fbion.close()
    ffln.close()
times=np.array(times)
```

```python
fformat0='%Y%m%d'
rivFluxPath='../../save/rivFluxesNoInterp'+t0.strftime(fformat0)+'-'+te.strftime(fformat0)+'.pkl'

pickle.dump((times,sumSilSOGRivers,sumSilFraser,sumNO3SOGRivers,sumNO3Fraser,sumNH4SOGRivers,sumNH4Fraser),
            open(rivFluxPath,'wb'))
```
